```python
import pandas as pd
import requests

url = 'https://earthquake.usgs.gov/fdsnws/event/1/query.csv'
```

```python
urls = []

for year in range(1900, 2023):
    for month in range(1, 13):
        if month < 10:
            month = '0' + str(month)
        else:
            month = str(month)
        if month in ['04', '06', '09', '11']:
            params = {'starttime': str(year) + '-' + month + '-01 00:00:00',
                      'endtime': str(year) + '-' + month + '-30 23:59:59',
                      'minmagnitude': '2.5',
                      'orderby': 'time-asc'}
        elif month == '02':
            if year % 4 == 0:
                params = {'starttime': str(year) + '-' + month + '-01 00:00:00',
                          'endtime': str(year) + '-' + month + '-29 23:59:59',
                          'minmagnitude': '2.5',
                          'orderby': 'time-asc'}
            else:
                params = {'starttime': str(year) + '-' + month + '-01 00:00:00',
                          'endtime': str(year) + '-' + month + '-28 23:59:59',
                          'minmagnitude': '2.5',
                          'orderby': 'time-asc'}
        else:
            params = {'starttime': str(year) + '-' + month + '-01 00:00:00',
                      'endtime': str(year) + '-' + month + '-31 23:59:59',
                      'minmagnitude': '2.5',
                      'orderby': 'time-asc'}
        name = str(year) + '-' + month + '.csv'
        urls.append((url, params, name))
```

```python
print(urls)
```

```python
from joblib import Parallel, delayed
def crawl(url, params, name):
    print(name)
    r = requests.get(url, params=params)
    with open('datatest/' + name, 'w') as f:
        f.write(r.text)
    return 0

Parallel(n_jobs=-1)(delayed(crawl)(url, params, name) for url, params, name in urls)
```
