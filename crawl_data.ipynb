```python
import pandas as pd
import requests

url = 'https://earthquake.usgs.gov/fdsnws/event/1/query.csv'
```

```python
# download data by month from 1900 to 2023

errs = []

for year in range(1900, 2023):
    for month in range(1, 13):
        if month < 10:
            month = '0' + str(month)
        else:
            month = str(month)
        if month in ['04', '06', '09', '11']:
            params = {'starttime': str(year) + '-' + month + '-01 00:00:00',
                      'endtime': str(year) + '-' + month + '-30 23:59:59',
                      'minmagnitude': '2.5',
                      'orderby': 'time-asc'}
        elif month == '02':
            if year % 4 == 0:
                params = {'starttime': str(year) + '-' + month + '-01 00:00:00',
                          'endtime': str(year) + '-' + month + '-29 23:59:59',
                          'minmagnitude': '2.5',
                          'orderby': 'time-asc'}
            else:
                params = {'starttime': str(year) + '-' + month + '-01 00:00:00',
                          'endtime': str(year) + '-' + month + '-28 23:59:59',
                          'minmagnitude': '2.5',
                          'orderby': 'time-asc'}
        else:
            params = {'starttime': str(year) + '-' + month + '-01 00:00:00',
                      'endtime': str(year) + '-' + month + '-31 23:59:59',
                      'minmagnitude': '2.5',
                      'orderby': 'time-asc'}
        r = requests.get(url, params=params)
        if r.status_code != 200:
            print("{}-{} is not downloaded".format(year, month))
            print(r.status_code)
            errs.append((year, month))
            
        else:
            print("{}-{} is downloaded, lines = {}".format(year, month, len(r.text.split('\n'))))
        with open('data/' + str(year) + '-' + month + '.csv', 'w') as f:
            f.write(r.text)

with open('data/errs.txt', 'w') as f:
    for err in errs:
        f.write(str(err[0]) + '-' + str(err[1]) + '\n')
```

```python
# conbine data into one csv file

import os

files = os.listdir('data')
files = [file for file in files if file.endswith('.csv')]
files.sort()

df = pd.read_csv('data/' + files[0])
print(df.head())
for file in files[1:]:
    print(file)
    df = pd.concat([df, pd.read_csv('data/' + file)])

df.to_csv('data/earthquake.csv', index=False)
```
